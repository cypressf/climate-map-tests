# Calculate summary statistics for toxic site flood data

```python
import pandas
import seaborn

seaborn.set_theme(style="whitegrid")

data = pandas.read_excel("ZipCode_Flooding.xlsx");

def between(df, min, max):
    return df[df["avg_risk_score_2_10"].between(min, max, inclusive="left")]

for i in range(1, 11):
    print(f"Risk {i-1} to {i}: {len(between(data, i - 1, i).index)}")

ax = seaborn.histplot(data["avg_risk_score_2_10"])
ax.set(xlabel="Flood Risk Score", ylabel="Number of toxic sites")
ax.set_title("Toxic Site Flood Risk Score Distribution")
```
