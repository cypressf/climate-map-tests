# Commands to convert .csv into .sql files

```bash
awk  '{print "("$0"),"}' file.csv > file.sql # add parens
sed 's/\(,,\)/d' file.sql > file.sql # remove rows with no data
sed 's/\([0-9][0-9][0-9][0-9]-[0-9][0-9]-[0-9][0-9]\)/"\1"/g' file.sql > file.sql # quote dates
split -l 200000 -d --additional-suffix=.sql file.sql file # split into smaller chunks so sqlx can run it
```

Then you have to go manually add the
```sql
INSERT INTO
    county_data (state_id,county_id,value,start_date,end_date,source,dataset)
VALUES
```
at the beginning of each file and the `;` at the end of each file.
