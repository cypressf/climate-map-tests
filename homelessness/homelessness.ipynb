## [Population totals](https://www.census.gov/data/datasets/time-series/demo/popest/2010s-state-total.html#par_textimage_500989927), using "Population Estimate 2019"

From the US Census

```python
homelessness = pandas.read_csv("input/homelessness.csv")
population = pandas.read_csv("input/population.csv")
state_county = pandas.read_csv("input/state_county.csv", dtype="str")
homelessness["STATEFP"] = homelessness["state"].apply(lambda state_name: us.states.lookup(state_name).fips)
population["STATEFP"] = population["state"].apply(lambda state_name: us.states.lookup(state_name).fips)
del homelessness["state"]
del population["state"]
data = pandas.merge(homelessness, population, on="STATEFP")
data["proportion_homeless_2019"] = data["homeless_people_2019"] / data["population_2019"]
del data["homeless_people_2019"]
del data["population_2019"]
data = pandas.merge(data, state_county, on="STATEFP")
data["proportion_homeless_2019"] = data["proportion_homeless_2019"].apply(lambda x: round(x, 6))
data.to_csv("output/homelessness.csv", index=None)
data
```
